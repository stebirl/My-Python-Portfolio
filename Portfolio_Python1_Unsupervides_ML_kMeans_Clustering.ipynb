```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Generate time axis
start_date = datetime(2024, 1, 1)
dates = [start_date + timedelta(hours=i) for i in range(1000)]

# Generate base patterns
t = np.linspace(0, 10, 1000)
sine_wave = np.sin(t)
cosine_wave = np.cos(t)
linear_trend = np.linspace(0, 1, 1000)
exponential_trend = np.exp(t/10) / np.exp(1)
square_wave = np.where(sine_wave >= 0, 1, -1)
sawtooth_wave = t % 1

# Generate variables with patterns and relationships
var1 = sine_wave + 0.1 * np.random.randn(1000)
var2 = cosine_wave + 0.1 * np.random.randn(1000)  # Anticorrelated with var1
var3 = linear_trend + 0.5 * sine_wave + 0.05 * np.random.randn(1000)  # Combination of linear trend and sine wave
var4 = exponential_trend + 0.3 * cosine_wave + 0.05 * np.random.randn(1000)  # Combination of exponential trend and cosine wave
var5 = -0.5 * var1 + 0.5 * var2 + 0.05 * np.random.randn(1000)  # Linear combination of var1 and var2
var6 = square_wave + 0.2 * np.random.randn(1000)  # Square wave pattern
var7 = sawtooth_wave + 0.1 * np.random.randn(1000)  # Sawtooth wave pattern
var8 = 0.5 * var3 + 0.5 * var4 - 0.3 * var6 + 0.05 * np.random.randn(1000)  # Complex combination of other variables

# Create DataFrame
df = pd.DataFrame({
    'Timestamp': dates,
    'Variable1': var1,
    'Variable2': var2,
    'Variable3': var3,
    'Variable4': var4,
    'Variable5': var5,
    'Variable6': var6,
    'Variable7': var7,
    'Variable8': var8
})

# Set Timestamp as index
df.set_index('Timestamp', inplace=True)

# Display first few rows and basic information
print(df.head())
print("\nDataset Info:")
print(df.info())

# Display correlation matrix
print("\nCorrelation Matrix:")
print(df.corr())

# Save to CSV (optional)
df.to_csv('patterned_dataset_8vars.csv')

# Plotting (optional, requires matplotlib)
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
for column in df.columns:
    plt.plot(df.index, df[column], label=column)
plt.legend()
plt.title('Variables Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.show()
```

```python
# Step 1: Scaling data and convert to array
from sklearn.preprocessing import scale
data = scale(df)
data
#from sklearn.decomposition import FactorAnalysis
```

```python
# Step 2: Estimate number of Factors
# Here we use simple Kaiser criterion (eigenvalues of data's covariance matrix over a threshold of 1)
# Outcome = 3 Factors
cov_mat = np.cov(data,rowvar=False)
ev = np.linalg.eig(cov_mat) [0]
plt.plot(ev,"o-")
plt.axhline(y=1.0,ls="--",color="red")
plt.xlabel("Eigenvalue ev")
plt.ylabel("Magnitude")
```

```python
## Step 3: Reduce dataset and by extracting factor loadings and plot
from sklearn.decomposition import FactorAnalysis
fa_n = 3 # Known by Kaiser Criterion
fa_model = FactorAnalysis(n_components = fa_n)
fa_model.fit(data)
factor_data = fa_model.transform(data)
loading = fa_model.components_
component_names = df.columns.values
loading_df = pd.DataFrame(loading, columns=component_names)
for i in range(fa_n):
    plt.barh(loading_df.columns,loading_df.iloc[i,:])
    plt.xlim(1,-1)
    plt.show()
```

```python
# Step 4: Clustering
# Here we make use of simple kMeans algorithm clustering
from sklearn.cluster import KMeans
inertias = []
n_clusters = []
maxK = 20
for k in range(1, maxK):
    kmeans = KMeans(n_clusters=k,n_init=10, random_state=None).fit(factor_data)
    n_clusters.append(k)
    inertias.append(kmeans.inertia_)
plt.plot(n_clusters, inertias, '.-')
```

```python
# Schritt 5: Cluster zum DF hinzufügen

DF = df.copy()
clu_n=3
km = KMeans(n_clusters=clu_n, random_state=None,n_init=10)
DF["km_clu"] = km.fit_predict(factor_data)
```
